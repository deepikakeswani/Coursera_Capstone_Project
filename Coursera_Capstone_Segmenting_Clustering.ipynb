<h3>Read table from wikipedia link into dataframe</h3>

```python
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
df
```

```python
#Delete Cells with Borough as Not assigned
new_df = df[df['Borough']!="Not assigned"]
```

```python
#Neighbourhood having value "Not assigned"
print(new_df[new_df['Neighbourhood']=='Not assigned'])
```

```python
#If Neighbourhood is Not assigned assign it value same as Borough
new_df.loc[new_df['Neighbourhood']=='Not assigned','Neighbourhood']=new_df['Borough']
```

```python
#Combine rows having same Postcode with Neighbourhood seperated by commas
new_df = new_df.groupby('Postcode').agg({'Borough':'first', 
                             'Neighbourhood': ', '.join, 
                             }).reset_index()
```
